```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/Ice_cream selling data.csv')
df.head()
```

```python
x=df.iloc[:,:-1]
x.ndim
```

```python
y=df.iloc[:,-1]
y
```

```python
plt.scatter(x,y)
plt.xlabel('Total Icecream Sales(Units)',fontweight='bold',color='r')
plt.ylabel('Temperature(Celsius)',fontweight='bold',color='r')
plt.title('Total Icecream Sale vs Temperature Graph',fontsize=20,fontweight='bold',color='k')
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
y_pred=model.predict(x)
y_pred
```

```python
print(model.predict([[2]]))
```

```python
df1 = pd.DataFrame({
    'Actual Value': y,
    'Predicted Value': y_pred,
    'Difference': y- y_pred
})
df1 = df1.reset_index(drop=True)  # Reset index without creating a new column for the old index
df1
```

```python
plt.scatter(x,y,color='b')
plt.plot(x,y_pred,color='r')
plt.xlabel('Total Icecream Sales(Units)',fontweight='bold',color='r')
plt.ylabel('Temperature(Celsius)',fontweight='bold',color='r')
plt.title('Total Icecream Sale vs Temperature Graph',fontsize=20,fontweight='bold',color='k')
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print('MAPE is',mean_absolute_percentage_error(y,y_pred))
print('R2 score is',r2_score(y,y_pred))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3)
x_poly=poly.fit_transform(x)
x_poly
```

```python
model1=LinearRegression()
model1.fit(x_poly,y)
y_poly=model1.predict(x_poly)
y_poly
```

```python
df1 = pd.DataFrame({
    'Actual Value': y,
    'Predicted Value': y_poly,
    'Difference': y- y_poly
})
df1 = df1.reset_index(drop=True)  # Reset index without creating a new column for the old index
df1
```

```python
print('MAPE is',mean_absolute_percentage_error(y,y_poly))
print('R2 score is',r2_score(y,y_poly))
```
